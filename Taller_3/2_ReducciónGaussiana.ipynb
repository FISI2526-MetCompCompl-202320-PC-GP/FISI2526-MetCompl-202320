```python
def gauss(A, b):
    n = len(A)
    x = [0] * n

    for i in range(n):
        p = i
        for j in range(i+1, n):
            if abs(A[j][i]) > abs(A[p][i]):
                p = j
        A[i], A[p] = A[p], A[i]
        b[i], b[p] = b[p], b[i]

        for j in range(i+1, n):
            A[j][i] /= A[i][i]
            b[j] -= A[j][i] * b[i]

    x[n-1] = b[n-1] / A[n-1][n-1]
    for i in range(n-2, -1, -1):
        x[i] = b[i]
        for j in range(i+1, n):
            x[i] -= A[i][j] * x[j]
        x[i] /= A[i][i]

    return x
```
